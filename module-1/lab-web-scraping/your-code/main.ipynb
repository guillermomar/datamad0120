####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
res = requests.get('https://www.emsc-csem.org/Earthquake/')
soup = BeautifulSoup(res.text, 'html.parser')
````

````python
def procesaFila(tag):
    m = tag.find_all("td")
    print(m[11].text.strip())
    return {
        "Date Time":m[3].text.strip(),
        "Latitude":m[4].text + " " + m[5].text,
        "Longitude":m[6].text.strip() + " " + m[7].text,
        "Region":m[11].text
    }

earthquake = soup.find_all('tbody')[0]

earthquake_dict = [procesaFila(fila) for fila in earthquake.find_all("tr")[1:]]
````

````python
df = pd.DataFrame(earthquake_dict)
df
````
