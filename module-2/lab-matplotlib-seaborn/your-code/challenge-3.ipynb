#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
df = pd.pivot_table(data=titanic,
                    index='Pclass',
                    values='PassengerId',
                    columns='Fare')
sns.heatmap(df,cmap='coolwarm')

#Esto es lo que hemos entendido, que había que hacer un heatmap con la correlación entre dos variables. Efectivente aquí vemos que 
# en la clase 1 están los clientes con precios más altos y viceversa
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('Correlation matrix')
```
